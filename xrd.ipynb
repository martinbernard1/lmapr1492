```python
# On import
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
```

```python
# On télécharge notre fichier CIF
with MPRester("3434NTJ0Daan9bDUnO") as m:
    structure = m.get_structure_by_material_id("mp-7914")
```

```python
# XRD
xrd = XRDCalculator('CuKa')
# Pattern
pattern = xrd.get_pattern(structure, scaled=True, two_theta_range=(0, 90))
```

```python
# On plot nos pics allant de 0 à 90 degrés en retirant les indices pour y voir claire
xrd.show_plot(structure, annotate_peaks=False)
```

```python
# On peut zoomer sur les 3 premiers pics en changeant les angles dans notre fonction
xrd.show_plot(structure, annotate_peaks=True, two_theta_range=(0, 33))
```

```python
# On peut donc maintenant bien observer les 3 premiers pics du diffractogramme 
# pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å

# On va maintenant calculer plus précisément l'intensité, l'angle et les indices hkl associés à chaque pic
```

```python
# Premier pic
print('Le premier pic se trouve en 2θ =',pattern.x[0] ,'°')
print('Son intensité est de',pattern.y[0] ,'%')
print('Indices hkl :',pattern.hkls[0])
```

```python
# Deuxième pic
print('Le deuxième pic se trouve en 2θ =',pattern.x[1] ,'°')
print('Son intensité est de',pattern.y[1] ,'%')
print('Indices hkl :',pattern.hkls[1])
```

```python
# Troisième pic
print('Le troisième pic se trouve en 2θ =',pattern.x[2] ,'°')
print('Son intensité est de',pattern.y[2] ,'%')
print('Indices hkl :',pattern.hkls[2])
```
